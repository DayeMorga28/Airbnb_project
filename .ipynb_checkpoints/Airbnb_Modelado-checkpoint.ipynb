```python
# Carga de librerias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_comentarios_merges = pd.read_csv('comentarios_merge.csv')
```

```python
df_comentarios_merges.dtypes
```

```python
# Cambiar tipo de datos
df_comentarios_merges['fecha_inicio_anfitrion'] = pd.to_datetime(df_comentarios_merges['fecha_inicio_anfitrion'])
df_comentarios_merges['Fecha'] = pd.to_datetime(df_comentarios_merges['Fecha'])
df_comentarios_merges['Anio'] = df_comentarios_merges['Anio'].astype(str)
df_comentarios_merges['num_habitaciones'] = df_comentarios_merges['num_habitaciones'].astype(int)
df_comentarios_merges['total_alojamientos_anfitrion'] = df_comentarios_merges['total_alojamientos_anfitrion'].astype(int)


#renombrar
df_comentarios_merges.rename(columns={'Fecha': 'fecha_comentario'}, inplace = True)
df_comentarios_merges
```

```python
# Graficación de alojamientos acumulados por ciudades
df_airbnb_sum_cities = df_comentarios_merges.groupby('ciudad').size()
df_airbnb_sum_cities.sort_values(ascending=False)
```

```python
# Graficación de alojamientos acumulados por ciudades
crosstab = pd.crosstab(df_comentarios_merges['ciudad'], df_comentarios_merges['superanfitrion'])
crosstab
```

```python
# Graficación de alojamientos acumulados por año de inicio anfitrion y superanfitrion
crosstab = pd.crosstab(df_comentarios_merges['Anio'], df_comentarios_merges['superanfitrion'])
crosstab
```

```python
# Graficación de alojamientos acumulados por tiempo de respuesta de anfitrion y superanfitrion
crosstab = pd.crosstab(df_comentarios_merges['tiempo_respuesta_anfitrion'], df_comentarios_merges['superanfitrion'])
crosstab
```

```python
# Graficación de alojamientos acumulados por ciudad y superanfitrion
crosstab = pd.crosstab(df_comentarios_merges['tasa_respuesta_anfitrion'], df_comentarios_merges['superanfitrion'])
crosstab
```

```python
# Graficación de identidad verificada del anfitrion y superanfitrion
crosstab = pd.crosstab(df_comentarios_merges['identidad_verificada_anfitrion'], df_comentarios_merges['superanfitrion'])
crosstab
```

```python
# Graficación de tipo de propiedad y superanfitrion
crosstab = pd.crosstab(df_comentarios_merges['tipo_propiedad'], df_comentarios_merges['superanfitrion'])
crosstab
```

```python
# Graficación de tipo de habitación y superanfitrion
crosstab = pd.crosstab(df_comentarios_merges['tipo_habitacion'], df_comentarios_merges['superanfitrion'])
crosstab
```

```python
# Graficación de reservación instantanea y superanfitrion
crosstab = pd.crosstab(df_comentarios_merges['reservacion_instantanea'], df_comentarios_merges['superanfitrion'])
crosstab
```

```python
# Graficación de calificación de exactitud del anuncio y superanfitrion
crosstab = pd.crosstab(df_comentarios_merges['calificacion_exactitud_anuncio'], df_comentarios_merges['superanfitrion'])
crosstab
```
