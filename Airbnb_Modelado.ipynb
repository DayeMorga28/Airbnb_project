```python
# Carga de librerias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_airbnb = pd.read_csv('alojamientos_limpio.csv', index_col=0)

df_airbnb.head()
```

```python
df_airbnb.dtypes
```

```python
# Cambio el tipo de dato de object a datetime
df_airbnb['fecha_inicio_anfitrion'] = pd.to_datetime(df_airbnb['fecha_inicio_anfitrion'])

df_airbnb['Anio'] = df_airbnb['Anio'].astype(str)
```

```python
# Graficación de alojamientos acumulados por ciudades
df_airbnb_sum_cities = df_airbnb.groupby('ciudad').size()
df_airbnb_sum_cities.sort_values(ascending=False)
```

```python
# Graficación de alojamientos acumulados por ciudades
crosstab = pd.crosstab(df_airbnb['ciudad'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de alojamientos acumulados por año de inicio anfitrion y superanfitrion
crosstab = pd.crosstab(df_airbnb['Anio'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de alojamientos acumulados por tiempo de respuesta de anfitrion y superanfitrion
crosstab = pd.crosstab(df_airbnb['tiempo_respuesta_anfitrion'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de alojamientos acumulados por ciudad y superanfitrion
crosstab = pd.crosstab(df_airbnb['tasa_respuesta_anfitrion'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de identidad verificada del anfitrion y superanfitrion
crosstab = pd.crosstab(df_airbnb['identidad_verificada_anfitrion'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de tipo de propiedad y superanfitrion
crosstab = pd.crosstab(df_airbnb['tipo_propiedad'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de tipo de habitación y superanfitrion
crosstab = pd.crosstab(df_airbnb['tipo_habitacion'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de reservación instantanea y superanfitrion
crosstab = pd.crosstab(df_airbnb['reservacion_instantanea'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de calificación de exactitud del anuncio y superanfitrion
crosstab = pd.crosstab(df_airbnb['calificacion_exactitud_anuncio'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de tipo de habitación y superanfitrion
crosstab = pd.crosstab(df_airbnb['calificacion_comentarios_limpieza'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de tipo de habitación y superanfitrion
crosstab = pd.crosstab(df_airbnb['calificacion_comentarios_comunicacion'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de tipo de habitación y superanfitrion
crosstab = pd.crosstab(df_airbnb['calificacion_comentarios_localizacion'], df_airbnb['superanfitrion'])
crosstab
```
