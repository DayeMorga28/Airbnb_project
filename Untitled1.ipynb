```python
# Carga de librerias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_airbnb = pd.read_csv('alojamientos_limpio.csv', index_col=0)

df_airbnb.head()
```

```python
# Graficación de alojamientos acumulados por ciudades
df_airbnb_sum_cities = df_airbnb.groupby('ciudad').size()
df_airbnb_sum_cities.sort_values(ascending=False)
```

```python
# Graficación de alojamientos acumulados por ciudad y superanfitrion
crosstab = pd.crosstab(df_airbnb['ciudad'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de alojamientos acumulados por fecha inicio anfitrion y superanfitrion
crosstab = pd.crosstab(df_airbnb['fecha_inicio_anfitrion'], df_airbnb['superanfitrion'])
crosstab
```

```python
# Graficación de alojamientos acumulados por ciudad y superanfitrion
crosstab = pd.crosstab(df_airbnb['total_alojamientos_anfitrion'], df_airbnb['superanfitrion'])
crosstab
```
